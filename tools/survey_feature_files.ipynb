# Survey the protobuf files used to create unit test cases

This uses parts of ``gherkinize`` to parse the file, so we can look for special cases, patterns, anomalies, and the like.

In this specific notebook, we're looking for test cases where there's no "value" attribute.
This appears to mean the test result is a boolean True value.

```python
import gherkinize
```

```python
from pathlib import Path
import logging
import contextlib

logging.getLogger().setLevel(logging.ERROR)
logging.getLogger("parso").setLevel(logging.ERROR)
print(logging.getLogger().getChildren())


config_path = Path.cwd() / "tags.toml"
config = gherkinize.Config(config_path)
# config.features
```

```python
count = 0
cel_spec_path = Path.cwd().parent.parent / "cel-spec" / "tests" / "simple" / "testdata"
for test_case in sorted(cel_spec_path.glob("*.textproto")):
    with contextlib.redirect_stderr(None):
        feature = gherkinize.Feature.from_text_proto(config, test_case)
    # print(feature.name)
    for section in feature.sections:
        # print(" ", section.name)
        for scenario in section.scenarios:
            if scenario.result.value is None:
                if scenario.steps[-1] != "Then none is None":
                    raise Exception("Not a 'none is None' case")
                print(
                    feature.name,
                    section.name,
                    scenario.name,
                )
                print("  ", scenario.source)
                print(
                    "  ",
                    [
                        f"{desc.name} = {value!r}"
                        for desc, value in scenario.source.ListFields()
                    ],
                )
                count += 1
print(f"{count} tests with where `value = True` is assumed, not stated.")
```
